### Preprocess data

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, '../../utils')
from ipynb.fs.defs.preprocessing import preprocess_dataset

X_train, X_test, y_train, y_test = preprocess_dataset(dataset_url = 'Salary_Data.csv')
```

### Train Linear Regression model on Training set

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

### Predict the Test set results

```python
y_predict = model.predict(X_test)
```

```python
print(y_predict)
print(y_test)
```

### Visualize the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Salary vs. Expericence (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Visualize the Test set results

```python
plt.scatter(X_test, y_test, color = 'red') # actual data
plt.scatter(X_test, y_predict, color = 'black') # predicted data
plt.plot(X_train, model.predict(X_train), color = 'blue') 
plt.title('Salary vs. Expericence (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Make a single prediction

```python
print('Predicted salary of an employee with 12 YOE:', model.predict([[12]])[0]) # input: 2D array, output: 1D array 1 element
```

### Get the equation of model

```python
print('Salary =', model.coef_[0], '* YOE +', model.intercept_)
```
